```python
from bs4 import BeautifulSoup
import requests
import time
```

```python
#https://www.ligue1.fr/classement?seasonId={saison}&StatsActiveTab=0&matchDay={journee}
```

```python
#saisons=["2013-2014","2014-2015","2015-2016","2016-2017","2017-2018","2018-2019","2019-2020","2020-2021","2021-2022","2022-2023"]

saisons=["1993-1994","1994-1995","1995-1996","1996-1997","1997-1998","1998-1999","1999-2000","2000-2001","2001-2002","2002-2003","2003-2004","2004-2005","2005-2006","2006-2007","2007-2008","2008-2009","2009-2010","2010-2011","2011-2012","2012-2013","2013-2014","2014-2015","2015-2016","2016-2017","2017-2018","2018-2019","2019-2020","2020-2021","2021-2022","2022-2023"]
for saison in saisons :
    for journee in range (5,39):
        #url=f'https://www.ligue1.fr/classement?matchDay={journee}'
        url=f'https://www.ligue1.fr/classement?seasonId={saison}&StatsActiveTab=0&matchDay={journee}'
        
        reponse=requests.get(url)
        soup=BeautifulSoup(reponse.text,"html.parser")
        time.sleep(3)

        list_equipe=[]
        list_position=[]
        list_points=[]
        list_gagne=[]
        list_nul=[]
        list_perd=[]
        list_buts=[]
        list_buts_enc=[]
        
        sante=[]


        recup_equipe=soup.findAll('span',{'class':'GeneralStats-clubName desktop-item'})
        recup_position=soup.findAll('div',{'class':'GeneralStats-item GeneralStats-item--position same'})
        recup_points=soup.findAll('div',{'class':'GeneralStats-item GeneralStats-item--points'})
        recup_stats=soup.findAll('div',{'class':'GeneralStats-item RankPage-mobileHide'})
        recup_forme=soup.findAll('div',{'class':'GeneralStats-item RankPage-mobileHide forme'})
        recup_nuls=soup.findAll('div',{'class':'GeneralStats-item RankPage-mobileHide'})



        test=str(recup_forme)
        t=test.split('span class="')

        liste_notes=[]
        notes=[]
        note=0
       
        for element in t:
            #print (element)
            
            if "circle win" in element:
                #print("win")
                note=5
                notes.append(note)
            elif "circle lose" in element:
                #print("lose")
                note=1
                notes.append(note)
            elif "circle draw" in element:
                #print("draw")
                note=3
                notes.append(note)

            
            if "</span>\n<span>-</span>" in element:
                note=4
                notes.append(note)

            if '</div>, <div class="GeneralStats-item RankPage-mobileHide forme">\n<span>-</span>' in element:
                #print(element)
                liste_notes.append(notes)
                notes=[]
                note=4
                notes.append(note)
            
            elif '</div>' in element:
                #print(element)
                liste_notes.append(notes)
                notes=[]
      

        list_forme=liste_notes


        position=1
        for ligne_equipe in recup_equipe:
            list_equipe.append(ligne_equipe.text)
            list_position.append(position)
            position=position+1



        for ligne_points in recup_points:
            list_points.append(ligne_points.text)
        i=0
        position=1
        for ligne_stats in recup_stats:
                if i==0:
                    list_gagne.append(ligne_stats.text)
                    i=i+1
                elif i==1:
                    list_nul.append(ligne_stats.text)
                    i=i+1
                elif i==2:
                    list_perd.append(ligne_stats.text)
                    i=i+1
                elif i==3:
                    list_buts.append(ligne_stats.text)
                    i=i+1
                elif i==4:
                    list_buts_enc.append(ligne_stats.text)
                    i=0




        #Traitement pour ajouter des 0 à list_forme pour les premieres journées
        if len(list_forme[0])==0:
            for i in range (0,len(list_forme)):
                list_forme[i].append('1')
        if len(list_forme[0])==1:
            for i in range (0,len(list_forme)):
                list_forme[i].append('1')
        if len(list_forme[0])==2:
            for i in range (0,len(list_forme)):
                list_forme[i].append('1')
        if len(list_forme[0])==3:
            for i in range (0,len(list_forme)):
                list_forme[i].append('1')
        if len(list_forme[0])==4:
            for i in range (0,len(list_forme)):
                list_forme[i].append('1')

        #Met à 10 l'état de santé de toutes les équipes
        #Si une équipe a des joueurs blessés baisser cette valeur pour l'équipe en question
        for i in range (0,20):
            sante.append(10)






        #Créatio du fichier csv
        with open(f"JourneeToutesSaisons.csv", "a",encoding="utf-16") as myfile:
            #Ecrit l'entete qu'une seule fois lors du premier passage
            if saison=="1993-1994" and journee==5:
                myfile.write("Team,Position,NbMatchGagne,NbMatchNul,NbMatchPerdu,NbButMis,NbButEnc,Forme5,Forme4,Forme3,Forme2,Forme1,Sante,Year,Journee,J+1\n")
                
            for i in range(len(list_equipe)):
                myfile.write(list_equipe[i]+','+str(list_position[i])+','+str(list_gagne[i])+','+str(list_nul[i])+','+str(list_perd[i])+','+str(list_buts[i])+','+str(list_buts_enc[i])+','+str(list_forme[i][0])+','+str(list_forme[i][1])+','+str(list_forme[i][2])+','+str(list_forme[i][3])+','+str(list_forme[i][4])+','+str(sante[i])+','+str(saison)+','+str(journee)+','+str(journee+1)+'\n')
        print('J',journee)
    print(saison)
```
