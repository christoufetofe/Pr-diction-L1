```python
def nettoyage_equipe(list_equipe):
    list_equipe_sortie=[]
    for equipe in list_equipe:

        if equipe=='\n\nOLYMPIQUE LYONNAIS\nOL\n':
            equipe='OLYMPIQUE LYONNAIS'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nRC STRASBOURG ALSACE\nSTRASBOURG\n':
            equipe='RC STRASBOURG ALSACE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nRC STRASBOURG ALSACE\nSTRASBOURG\n\n':
            equipe='RC STRASBOURG ALSACE'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nCLERMONT FOOT 63\nCLERMONT\n':
            equipe='CLERMONT FOOT 63'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nCLERMONT FOOT 63\nCLERMONT\n\n':
            equipe='CLERMONT FOOT 63'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nTOULOUSE FC\nTOULOUSE FC\n':
            equipe='TOULOUSE FC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nTOULOUSE FC\nTOULOUSE FC\n\n':
            equipe='TOULOUSE FC'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nANGERS SCO\nANGERS SCO\n':
            equipe='ANGERS SCO'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nANGERS SCO\nANGERS SCO\n\n':
            equipe='ANGERS SCO'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nRC LENS\nRC LENS\n':
            equipe='RC LENS'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nRC LENS\nRC LENS\n\n':
            equipe='RC LENS'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nLOSC LILLE\nLOSC\n':
            equipe='LOSC LILLE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nLOSC LILLE\nLOSC\n\n':
            equipe='LOSC LILLE'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nMONTPELLIER HÉRAULT SC\nMONTPELLIER\n':
            equipe='MONTPELLIER HÉRAULT SC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nMONTPELLIER HÉRAULT SC\nMONTPELLIER\n\n':
            equipe='MONTPELLIER HÉRAULT SC'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nSTADE RENNAIS FC\nRENNES\n':
            equipe='STADE RENNAIS FC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nSTADE RENNAIS FC\nRENNES\n\n':
            equipe='STADE RENNAIS FC'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nOLYMPIQUE DE MARSEILLE\nOM\n':
            equipe='OLYMPIQUE DE MARSEILLE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nOLYMPIQUE DE MARSEILLE\nOM\n\n':
            equipe='OLYMPIQUE DE MARSEILLE'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nAC AJACCIO\nAC AJACCIO\n':
            equipe='AC AJACCIO'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nAC AJACCIO\nAC AJACCIO\n\n':
            equipe='AC AJACCIO'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nAS MONACO\nAS MONACO\n':
            equipe='AS MONACO'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nAS MONACO\nAS MONACO\n\n':
            equipe='AS MONACO'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nPARIS SAINT-GERMAIN\nPSG\n':
            equipe='PARIS SAINT-GERMAIN'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nPARIS SAINT-GERMAIN\nPSG\n\n':
            equipe='PARIS SAINT-GERMAIN'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nOGC NICE\nOGC NICE\n':
            equipe='OGC NICE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nOGC NICE\nOGC NICE\n\n':
            equipe='OGC NICE'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nFC NANTES\nFC NANTES\n':
            equipe='FC NANTES'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nFC NANTES\nFC NANTES\n\n':
            equipe='FC NANTES'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nSTADE BRESTOIS 29\nBREST\n':
            equipe='STADE BRESTOIS 29'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nSTADE BRESTOIS 29\nBREST\n\n':
            equipe='STADE BRESTOIS 29'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nAJ AUXERRE\nAJ AUXERRE\n':
            equipe='AJ AUXERRE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nAJ AUXERRE\nAJ AUXERRE\n\n':
            equipe='AJ AUXERRE'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nESTAC TROYES\nTROYES\n':
            equipe='ESTAC TROYES'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nESTAC TROYES\nTROYES\n\n':
            equipe='ESTAC TROYES'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nFC LORIENT\nFC LORIENT\n':
            equipe='FC LORIENT'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nFC LORIENT\nFC LORIENT\n\n':
            equipe='FC LORIENT'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nSTADE DE REIMS\nREIMS\n':
            equipe='STADE DE REIMS'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nSTADE DE REIMS\nREIMS\n\n':
            equipe='STADE DE REIMS'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nFC GIRONDINS DE BORDEAUX\nBORDEAUX\n':
            equipe='FC GIRONDINS DE BORDEAUX'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nFC GIRONDINS DE BORDEAUX\nBORDEAUX\n\n':
            equipe='FC GIRONDINS DE BORDEAUX'
            list_equipe_sortie.append(equipe)


        elif equipe=='\n\nFC METZ\nFC METZ\n':
            equipe='FC METZ'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nFC METZ\nFC METZ\n\n':
            equipe='FC METZ'
            list_equipe_sortie.append(equipe)


        elif equipe=='\n\nS.C. BASTIAIS\nSC BASTIA\n':
            equipe='S.C. BASTIAIS'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nS.C. BASTIAIS\nSC BASTIA\n\n':
            equipe='S.C. BASTIAIS'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nAS SAINT-ÉTIENNE\nASSE\n':
            equipe='AS SAINT-ÉTIENNE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nAS SAINT-ÉTIENNE\nASSE\n\n':
            equipe='AS SAINT-ÉTIENNE'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nHAVRE AC\nHAVRE AC\n':
            equipe='HAVRE AC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nHAVRE AC\nHAVRE AC\n\n':
            equipe='HAVRE AC'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nC.S. SEDAN ARDENNES\nCS Sedan\n':
            equipe='C.S. SEDAN ARDENNES'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nC.S. SEDAN ARDENNES\nCS Sedan\n\n':
            equipe='C.S. SEDAN ARDENNES'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nAS NANCY LORRAINE\nNANCY\n':
            equipe='AS NANCY LORRAINE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nAS NANCY LORRAINE\nNANCY\n\n':
            equipe='AS NANCY LORRAINE'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nEN AVANT GUINGAMP\nEA GUINGAMP\n':
            equipe='EN AVANT GUINGAMP'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nEN AVANT GUINGAMP\nEA GUINGAMP\n\n':
            equipe='EN AVANT GUINGAMP'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nFC SOCHAUX-MONTBÉLIARD\nFCSM\n':
            equipe='FC SOCHAUX-MONTBÉLIARD'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nFC SOCHAUX-MONTBÉLIARD\nFCSM\n\n':
            equipe='FC SOCHAUX-MONTBÉLIARD'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nLE MANS FC\nLE MANS FC\n':
            equipe='LE MANS FC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nLE MANS FC\nLE MANS FC\n\n':
            equipe='LE MANS FC'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nSTADE MALHERBE CAEN\nSM CAEN\n':
            equipe='STADE MALHERBE CAEN'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nSTADE MALHERBE CAEN\nSM CAEN\n\n':
            equipe='STADE MALHERBE CAEN'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nF.C. ISTRES OUEST PROVENCE\nFC Istres\n':
            equipe='F.C. ISTRES OUEST PROVENCE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nF.C. ISTRES OUEST PROVENCE\nFC Istres\n\n':
            equipe='F.C. ISTRES OUEST PROVENCE'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nVALENCIENNES FC\nVAFC\n':
            equipe='VALENCIENNES FC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nVALENCIENNES FC\nVAFC\n\n':
            equipe='VALENCIENNES FC'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nGRENOBLE FOOT 38\nGRENOBLE\n':
            equipe='GRENOBLE FOOT 38'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nGRENOBLE FOOT 38\nGRENOBLE\n\n':
            equipe='GRENOBLE FOOT 38'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nA.C. ARLES - AVIGNON\nAC Arles-Avignon\n':
            equipe='A.C. ARLES - AVIGNON'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nA.C. ARLES - AVIGNON\nAC Arles-Avignon\n\n':
            equipe='A.C. ARLES - AVIGNON'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nEVIAN THONON GAILLARD F. C.\nEvian TG\n':
            equipe='EVIAN THONON GAILLARD F. C.'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nEVIAN THONON GAILLARD F. C.\nEvian TG\n\n':
            equipe='EVIAN THONON GAILLARD F. C.'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nDIJON FCO\nDIJON FCO\n':
            equipe='DIJON FCO'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nDIJON FCO\nDIJON FCO\n\n':
            equipe='DIJON FCO'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nGAZELEC FC AJACCIO\nGAZ. AJACCIO\n':
            equipe='GAZELEC FC AJACCIO'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nGAZELEC FC AJACCIO\nGAZ. AJACCIO\n\n':
            equipe='GAZELEC FC AJACCIO'
            list_equipe_sortie.append(equipe)




        elif equipe=='\n\nAMIENS SC\nAMIENS SC\n':
            equipe='AMIENS SC'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nAMIENS SC\nAMIENS SC\n\n':
            equipe='AMIENS SC'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nNÎMES OLYMPIQUE\nNÎMES\n':
            equipe='NÎMES OLYMPIQUE'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nNÎMES OLYMPIQUE\nNÎMES\n\n':
            equipe='NÎMES OLYMPIQUE'
            list_equipe_sortie.append(equipe)



        elif equipe=="\n\nU.S. BOULOGNE COTE D'OPALE\nUSBCO\n":
            equipe="U.S. BOULOGNE COTE D'OPALE"
            list_equipe_sortie.append(equipe)
        elif equipe=="\n\n\nU.S. BOULOGNE COTE D'OPALE\nUSBCO\n\n":
            equipe="U.S. BOULOGNE COTE D'OPALE"
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nA.S. CANNES\nAS Cannes\n':
            equipe='A.S. CANNES'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nA.S. CANNES\nAS Cannes\n\n':
            equipe='A.S. CANNES'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nF.C. MARTIGUES\nFCM\n':
            equipe='F.C. MARTIGUES'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nF.C. MARTIGUES\nFCM\n\n':
            equipe='F.C. MARTIGUES'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nF.C. GUEUGNON\nFCG\n':
            equipe='F.C. GUEUGNON'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nF.C. GUEUGNON\nFCG\n\n':
            equipe='F.C. GUEUGNON'
            list_equipe_sortie.append(equipe)



        elif equipe=='\n\nCHÂTEAUROUX\nCHÂTEAUROUX\n':
            equipe='CHÂTEAUROUX'
            list_equipe_sortie.append(equipe)
        elif equipe=='\n\n\nCHÂTEAUROUX\nCHÂTEAUROUX\n\n':
            equipe='CHÂTEAUROUX'
            list_equipe_sortie.append(equipe)





        elif equipe=='\n\n\nOLYMPIQUE LYONNAIS\nOL\n\n':
            equipe='OLYMPIQUE LYONNAIS'
            list_equipe_sortie.append(equipe)




    return(list_equipe_sortie)




#liste des différentes saisons
saisons=["1993-1994","1994-1995","1995-1996","1996-1997","1997-1998","1998-1999","1999-2000","2000-2001","2001-2002","2002-2003","2003-2004","2004-2005","2005-2006","2006-2007","2007-2008","2008-2009","2009-2010","2010-2011","2011-2012","2012-2013","2013-2014","2014-2015","2015-2016","2016-2017","2017-2018","2018-2019","2019-2020","2020-2021","2021-2022","2022-2023"]
for saison in saisons :
    #On commence le scraping à partir de la 6eme 
    for journee in range (6,39):
        url=f'https://www.ligue1.fr/calendrier-resultats?seasonId={saison}&matchDay={journee}&StatsActiveTab=0'
        reponse=requests.get(url)
        soup=BeautifulSoup(reponse.text,"html.parser")
        equipe_home=[]
        equipe_away=[]
        list_equipe_home=[]
        list_equipe_away=[]
        scoreatraiter=[]
        score_home=[]
        score_away=[]
        result=[]

        #Scraping de la liste des équipe hôtes
        recup_equipe_home=soup.findAll('div',{'class':'Calendar-clubWrapper Calendar-clubWrapper--left'})
        #Scraping de la liste des équipe adveairse
        recup_equipe_away=soup.findAll('div',{'class':'club away'})
        #Récup du score
        recup_score=soup.findAll('div',{'class':'Calendar-clubResult result'})

        #Traitement du score
        for ligne_score in recup_score:
            scoreatraiter.append(ligne_score.text)
        for texte_score in scoreatraiter:
            if texte_score=="\n\nReporté\n\n":
                texte_score="\n\n99-99\n\n"
            elif texte_score=="\n\nAnnulé\n\n":
                texte_score="\n\n99-99\n\n"
            if ":" in texte_score:
                break
            
            score_home.append(texte_score.split('\n\n')[1].split('-')[0])
            score_away.append(texte_score.split('\n\n')[1].split('-')[1])
        
        for ligne_equipe in recup_equipe_home:
            equipe_home.append(ligne_equipe.text)
        for ligne_equipe in recup_equipe_away:
            equipe_away.append(ligne_equipe.text)

        #créer la variable sortie en fonction du score 
        #0 pour match nul 
        #1 pour victoire de l'equipe home
        #2 pour victoire de l'équipe away
        for i in range (0,len(score_home)):
            if score_home[i]==score_away[i]:
                result.append(0)
            elif score_home[i]>score_away[i]:
                result.append(1)
            elif score_home[i]<score_away[i]:
                result.append(2)


        #Nettoyage de la liste des équipes hôte et advairse
        list_equipe_away=nettoyage_equipe(equipe_away)
        list_equipe_home=nettoyage_equipe(equipe_home)


        #Enregistrement dans un fichier CSV
        with open(f"CalendrierToutesSaisons.csv", "a",encoding="utf-16") as myfile:
            #Ecrit l'entete qu'une seule fois lors du premier passage
            if saison=="1993-1994" and journee==6:
                myfile.write("home_team,home_score,away_score,away_team,result,saison,Journee\n")            
            for i in range(len(score_home)):
                myfile.write(list_equipe_home[i]+','+str(score_home[i])+','+str(score_away[i])+','+(list_equipe_away[i])+','+str(result[i])+','+str(saison)+','+str(journee)+'\n')
    print(saison)
```
